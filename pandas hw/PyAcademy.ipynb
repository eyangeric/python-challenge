```python
#import modules
import pandas as pd
import os
```

```python
#link to datasets and read them in
schools_complete_filepath = os.path.join("..", "..", "..", "Desktop", "NUCHI201811DATA2", "Homework", "04-Pandas", "Instructions", "PyCitySchools", "Resources", "schools_complete.csv")
students_complete_filepath = os.path.join("..", "..", "..", "Desktop", "NUCHI201811DATA2", "Homework", "04-Pandas", "Instructions", "PyCitySchools", "Resources", "students_complete.csv")
```

```python
#read in datasets
schools_complete = pd.read_csv(schools_complete_filepath)
students_complete = pd.read_csv(students_complete_filepath)
```

```python
#check datasets
print(schools_complete.head())
print(students_complete.head())
print(students_complete.columns)
```

```python
####################################District Summary######################################
total_schools = len(students_complete["school_name"].unique())
total_students = len(students_complete["Student ID"].unique())
total_budget = schools_complete["budget"].sum()
average_math_score = students_complete["math_score"].mean()
average_reading_score = students_complete["reading_score"].mean()
percent_passing_math = len(students_complete.loc[students_complete["math_score"] >= 70, "Student ID"].unique()) / len(students_complete["Student ID"].unique()) * 100
percent_passing_reading = len(students_complete.loc[students_complete["reading_score"] >= 70, "Student ID"].unique()) / len(students_complete["Student ID"].unique()) * 100
overall_passing_rate = (average_math_score + average_reading_score)/2

District_Dict = {"Total Schools": total_schools, "Total Students": total_students,
                                 "Total Budget": total_budget, "Average Math Score": average_math_score, 
                                 "Average Reading Score": average_reading_score,
                                 "% Passing Math": percent_passing_math, "% Passing Reading": percent_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate}
District_Summary = pd.DataFrame(District_Dict, columns=["Total Schools", "Total Students", 
                                                        "Total Budget", "Average Math Score", 
                                                        "Average Reading Score", "% Passing Math",
                                                        "% Passing Reading", "% Overall Passing Rate"],
                                                        index=[0])
```

```python
#format District Summary variables
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
#convert Total Budget to float before formatting
District_Summary["Total Budget"] = District_Summary["Total Budget"].astype(float)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)

#Print District Summary
District_Summary
```

```python
####################################School Summary######################################
students_schoolgroup = students_complete.groupby(["school_name"])
students_schoolgroup.count()
grouped_avg_math_score = students_schoolgroup["math_score"].mean()
grouped_avg_reading_score = students_schoolgroup["reading_score"].mean()
students_schoolgroup["math_score"].count()



#set schools data index to school name
schools_complete = schools_complete.set_index("school_name")
#sort index (school name) by ascending alphabetical order
schools_complete = schools_complete.sort_index(axis=0, ascending=True)
```

```python
#restrict to math/reading scores >= 70 before using groupby statement
students_complete_math70 = students_complete.loc[students_complete["math_score"] >= 70, :]
students_complete_reading70 = students_complete.loc[students_complete["reading_score"] >= 70, :]

#group by School
students_schoolgroup_math70 = students_complete_math70.groupby(["school_name"])
students_schoolgroup_reading70 = students_complete_reading70.groupby(["school_name"])
math_pass_school = students_schoolgroup_math70["math_score"].count()
reading_pass_school = students_schoolgroup_reading70["reading_score"].count()
```

```python
#merge school grouped dataframes
School_Summary = pd.DataFrame({"School Type": schools_complete["type"], "Total Students": schools_complete["size"], 
                               "Total School Budget": schools_complete["budget"], 
                               "Per Student Budget": schools_complete["budget"]/schools_complete["size"],
                               "Average Math Score": grouped_avg_math_score, "Average Reading Score": grouped_avg_reading_score,
                               "% Passing Math": math_pass_school/schools_complete["size"]*100, 
                               "% Passing Reading": reading_pass_school/schools_complete["size"]*100,
                               "Overall Passing Rate": ((math_pass_school/schools_complete["size"]*100)+(reading_pass_school/schools_complete["size"]*100))/2 })

#format variables
School_Summary["Total School Budget"] = School_Summary["Total School Budget"].map("${:,.2f}".format)
School_Summary["Per Student Budget"] = School_Summary["Per Student Budget"].map("${:,.2f}".format)

#print School Summary
School_Summary
```

```python
####################################Top Performing Schools (By Passing Rate)######################################
Ordered_HightoLow_School_Summary = School_Summary.sort_values(by='Overall Passing Rate', ascending=False)
Top5_School_Summary = Ordered_HightoLow_School_Summary.head()

#print top 5 
Top5_School_Summary
```

```python
####################################Bottom Performing Schools (By Passing Rate)######################################
Ordered_LowtoHigh_School_Summary = School_Summary.sort_values(by='Overall Passing Rate')
Bottom5_School_Summary = Ordered_LowtoHigh_School_Summary.head()

#print top 5 
Bottom5_School_Summary
```

```python
####################################Math Scores by Grade######################################
#restrict by grade in separate datasets
Grade9 = students_complete[students_complete["grade"] == "9th"]
Grade10 = students_complete[students_complete["grade"] == "10th"]
Grade11 = students_complete[students_complete["grade"] == "11th"]
Grade12 = students_complete[students_complete["grade"] == "12th"]

#group each grade subset by school name
Grade9_schoolgroup = Grade9.groupby('school_name')
Grade10_schoolgroup = Grade10.groupby('school_name')
Grade11_schoolgroup = Grade11.groupby('school_name')
Grade12_schoolgroup = Grade12.groupby('school_name')

#calculate means for each grade grouped by schools
Grade9_schoolgroup_math = Grade9_schoolgroup["math_score"].mean()
Grade10_schoolgroup_math = Grade10_schoolgroup["math_score"].mean()
Grade11_schoolgroup_math = Grade11_schoolgroup["math_score"].mean()
Grade12_schoolgroup_math = Grade12_schoolgroup["math_score"].mean()

#create dataframe for math scores by grade and school
MathScores_Grade_School = pd.DataFrame({"9th": Grade9_schoolgroup_math, 
                                        "10th": Grade10_schoolgroup_math,
                                        "11th": Grade11_schoolgroup_math,
                                        "12th": Grade12_schoolgroup_math})

#print average math scores by grade and school
MathScores_Grade_School
```

```python
####################################Reading Scores by Grade######################################
#calculate means for each grade grouped by schools
Grade9_schoolgroup_reading = Grade9_schoolgroup["reading_score"].mean()
Grade10_schoolgroup_reading = Grade10_schoolgroup["reading_score"].mean()
Grade11_schoolgroup_reading = Grade11_schoolgroup["reading_score"].mean()
Grade12_schoolgroup_reading = Grade12_schoolgroup["reading_score"].mean()

#create dataframe for math scores by grade and school
ReadingScores_Grade_School = pd.DataFrame({"9th": Grade9_schoolgroup_reading, 
                                        "10th": Grade10_schoolgroup_reading,
                                        "11th": Grade11_schoolgroup_reading,
                                        "12th": Grade12_schoolgroup_reading})

#print average math scores by grade and school
ReadingScores_Grade_School
```

```python
#################################Scores by School Spending######################################

#create average spending per student variable in the original schools dataset
schools_complete["Per Student Budget"] = schools_complete["budget"] / schools_complete["size"]
#schools_complete["Per Student Budget"].describe()

#create bins to group school spending
#did not use my own bins because I wanted to compare the output with the results :) 
#If there were no example bins, I would've used 25th, 50th, and 75th percentiles to decide my bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#create average spending (per student) categories based on the bins
schools_complete["Per Student Budget Category"] = pd.cut(schools_complete["Per Student Budget"], spending_bins, labels=group_names)
```

```python
#create spendnig range variable in School Summary dataframe
School_Summary['Spending Ranges (Per Student)'] = schools_complete['Per Student Budget Category']

#group by average spending (per student) categories
School_Spending_Summary = School_Summary.groupby('Spending Ranges (Per Student)').mean()

#remove column of total students
del School_Spending_Summary['Total Students']

#show results score statistics by spending range categories
School_Spending_Summary
```

```python
#################################Scores by School Size######################################
#bins for school size
size_bins = [0, 1000, 2000, 5000]
size_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#create school size categories based on the bins
schools_complete["School Size"] = pd.cut(schools_complete["size"], size_bins, labels=size_group_names)

School_Summary['School Size'] = schools_complete['School Size']

#group by school size categories
School_Size_Summary = School_Summary.groupby('School Size').mean()

#remove column of total students
del School_Size_Summary['Total Students']

#show results score statistics by school size categories
School_Size_Summary
```

```python
#################################Scores by School Type######################################
#group by school type
School_Type_Summary = School_Summary.groupby('School Type').mean()

#remove column of total students
del School_Type_Summary['Total Students']

#show results score statistics by school type
School_Type_Summary
```
